### **1. A board has n*m cells, and there is a gift with some value (value is greater than 0) in every cell. You can get gifts starting from the top-left cell, and move right or down in each step, and finally reach the cell at the bottom-right cell. Given 2 numbers N and M followed by N*M numbers, What’s the maximal value of gifts you can get from the board?**

Input Size : 1 <= N <= 1000

**Sample Testcases:**

**INPUT**

3 3

4 0 0

0 4 0

0 0 4

**OUTPUT**

12

```python
def max_gift_value(N, M, board):
    dp = [[0] * M for _ in range(N)]
    dp[0][0] = board[0][0]
    
    for j in range(1, M):
        dp[0][j] = dp[0][j-1] + board[0][j]
    
    for i in range(1, N):
        dp[i][0] = dp[i-1][0] + board[i][0]

    for i in range(1, N):
        for j in range(1, M):
            dp[i][j] = max(dp[i-1][j], dp[i][j-1]) + board[i][j]
    return dp[N-1][M-1]

N, M = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(N)]
print(max_gift_value(N, M, board))
```

### **3. Given 2 numbers N,M where N is the number of nodes and M is the number of edges,followed by the connection between nodes, find if the resultant graph has cycles.(all the edges are directed edges).**

Input Size : M,N <= 100000

**Sample Testcase:**

**INPUT**

2 2

1 2

2 1

**OUTPUT**

yes

```python
def has_cycle(n, m, edges):
    graph = [[] for _ in range(n + 1)]
    for u, v in edges:
        graph[u].append(v)
    
    visited = [0] * (n + 1)
    
    def dfs(node):
        if visited[node] == 1:
            return True
        if visited[node] == 2:
            return False
        visited[node] = 1
        for neighbor in graph[node]:
            if dfs(neighbor):
                return True
        visited[node] = 2
        return False

    for i in range(1, n + 1):
        if visited[i] == 0:
            if dfs(i):
                return "yes"
    return "no"

n, m = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(m)]
print(has_cycle(n, m, edges))
```

### **4. Given 3 numbers N,E,A where N is the number of nodes and E is the number of edges(weight of each edge is 1), find the shortest path from node 1 to given node A.**

Input Size : E,N <= 100000

**Sample Testcase:**

**INPUT**

2 2 2

1 2

2 1

**OUTPUT**

1

```python
from collections import deque, defaultdict

def shortest_path(n, e, a, edges):
    graph = defaultdict(list)
    
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    
    distance = [-1] * (n + 1)
    queue = deque([1])       
    
    distance[1] = 0       
    while queue:
        node = queue.popleft()
        for neighbor in graph[node]:
            if distance[neighbor] == -1:
                distance[neighbor] = distance[node] + 1
                queue.append(neighbor)
    return distance[a]

n, e, a = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(e)]
print(shortest_path(n, e, a, edges))
```

### **5. You are given with number of nodes(N) and edges(E) in a tree followed by ‘E’ pairs of parent and child values,find whether it forms the binary tree structure print 'yes' otherwise print 'no'.**

Note:Let first value in the first pair be the root.

Child can be considered as either left or right child of the corresponding 
parent

Input Size : 1 <= N, M <= 100000

**Sample Testcase:**

**INPUT**

7 6

26 10

26 3

10 4

10 6

3 1

3 2

**OUTPUT**

yes

```python
from collections import defaultdict, deque

def is_binary_tree(N, E, edges):
    child_count = defaultdict(int)
    parent_count = defaultdict(int)
    adj_list = defaultdict(list)
    
    for parent, child in edges:
        adj_list[parent].append(child)
        child_count[parent] += 1
        parent_count[child] += 1
        
        if child_count[parent] > 2:
            return "no"
    
    root = None
    for node in range(1, N + 1):
        if parent_count[node] == 0:
            if root is None:
                root = node
            else:
                return "no"
    
    if root is None:
        return "no"
        
    visited = set()
    queue = deque([root])
    visited.add(root)
    
    while queue:
        current = queue.popleft()
        for neighbor in adj_list[current]:
            if neighbor in visited:
                return "no"
            visited.add(neighbor)
            queue.append(neighbor)
    
    if len(visited) != N:
        return "no"
    return "yes"

N, E = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(E)]
print(is_binary_tree(N, E, edges))
```

### **6. Vishal is learning to skate on ice. He's a beginner, so his only mode of transportation is pushing off from a snow drift to the north, east, south or west and sliding until he lands in another snow drift. He has noticed that in this way it's impossible to get from some snow drifts to some other by any sequence of moves. He now wants to heap up some additional snow drifts, so that he can get from any snow drift to any other one. He asked you to find the minimal number of snow drifts that need to be created. We assume that Vishal can only heap up snow drifts at integer coordinates. The first line of input contains a single integer N the number of snow drifts. Each of the following n lines contains two integers xi and yi the coordinates of the i-th snow drift. Note that the north direction coinсides with the direction of Oy axis, so the east direction coinсides with the direction of the Ox axis. All snow drift's locations are distinct.**

Input Size : 1 <= N,Xi,Yi <= 1000

**Sample Testcase:**

**INPUT**

2

2 1

1 2

**OUTPUT**

1

```python
def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, rank, x, y):
    rootX = find(parent, x)
    rootY = find(parent, y)
    if rootX != rootY:
        if rank[rootX] > rank[rootY]:
            parent[rootY] = rootX
        elif rank[rootX] < rank[rootY]:
            parent[rootX] = rootY
        else:
            parent[rootY] = rootX
            rank[rootX] += 1

def minimal_additional_snow_drifts(n, coordinates):
    parent = {i: i for i in range(n)}
    rank = {i: 0 for i in range(n)}
    x_map = {}
    y_map = {}
    for i, (x, y) in enumerate(coordinates):
        if x in x_map:
            union(parent, rank, i, x_map[x])
        x_map[x] = i
        if y in y_map:
            union(parent, rank, i, y_map[y])
        y_map[y] = i

    components = len(set(find(parent, i) for i in range(n)))
    return components - 1

n = int(input())
coordinates = [tuple(map(int, input().split())) for _ in range(n)]
print(minimal_additional_snow_drifts(n, coordinates))
```

### **7. Given 2 numbers N,M where N is the number of nodes and M is the number of edges, find if the resultant graph is a tree.**

Input Size : M,N <= 100000

**Sample Testcase:**

**INPUT**

3 2

1 2

2 3

**OUTPUT**

yes

```python
from collections import defaultdict, deque

def is_tree(N, M, edges):
    if M != N - 1:
        return "no"
    
    adj_list = defaultdict(list)
    for u, v in edges:
        adj_list[u].append(v)
        adj_list[v].append(u)
    
    visited = set()
    queue = deque([1])
    visited.add(1)
    parent = {1: -1}
    
    while queue:
        current = queue.popleft()
        for neighbor in adj_list[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                queue.append(neighbor)
            elif parent[current] != neighbor:
                return "no"
    if len(visited) == N:
        return "yes"
    else:
        return "no"

N, M = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(M)]
print(is_tree(N, M, edges))
```

### **8. Given a Binary Tree with N nodes and M edges, write a program to find a binary tree is symmetric or not. If it is symmetric print 'yes' else 'no'.**

Input Size : 2 <= N <= 100000

**Sample Testcases:**

**INPUT**

3 2

1 2

1 2

**OUTPUT**

YES

```python
from collections import defaultdict, deque

def is_symmetric_tree(N, M, edges):
    if N == 1:
        return "YES"

    adj_list = defaultdict(list)
    for u, v in edges:
        adj_list[u].append(v)
    for node in adj_list:
        adj_list[node].sort()
    
    def is_mirror(left_subtree, right_subtree):
        if not left_subtree and not right_subtree:
            return True
        if not left_subtree or not right_subtree:
            return False
        if len(left_subtree) != len(right_subtree):
            return False
        for l, r in zip(left_subtree, reversed(right_subtree)):
            if l != r or not is_mirror(adj_list[l], adj_list[r]):
                return False
        return True

    root = 1
    children = adj_list[root]
    if len(children) != 2:
        return "NO"
    return "YES" if is_mirror(adj_list[children[0]], adj_list[children[1]]) else "NO"

N, M = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(M)]
print(is_symmetric_tree(N, M, edges))
```

### **10. Given a number N- (number of nodes) followed by N numbers. Construct a BST with the inputs in the same order. Now 2 numbers U and V are given. Find the LCA( Lowest Common Anscestor) of U and V.**

Input Size : 1 <= N <= 10000

**Sample Testcases:**

**INPUT**

5

1 2 3 4 5

4 5

**OUTPUT**

4

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Insert a node into the BST
def insert_bst(root, value):
    if not root:
        return TreeNode(value)
    if value < root.value:
        root.left = insert_bst(root.left, value)
    else:
        root.right = insert_bst(root.right, value)
    return root

# Find the LCA iteratively
def find_lca(root, u, v):
    while root:
        if u < root.value and v < root.value:
            root = root.left
        elif u > root.value and v > root.value:
            root = root.right
        else:
            return root.value
    return -1

n = int(input())
nodes = list(map(int, input().split()))
u, v = map(int, input().split())

root = None
for value in nodes:
    root = insert_bst(root, value)
print(find_lca(root, u, v))
```

HINT : USE SEGMENT TREE

Input Size : N<=100000

**Example:**

**INPUT**

1 1 1 1 1

1 3

**OUTPUT**

```python
class SegmentTree:
    def __init__(self, arr):
        n = len(arr)
        self.n = n
        self.tree = [0] * (4 * n)
        self._build(arr, 0, 0, n - 1)
    
    def _build(self, arr, node, start, end):
        if start == end:
            self.tree[node] = arr[start]
        else:
            mid = (start + end) // 2
            left_child = 2 * node + 1
            right_child = 2 * node + 2
            self._build(arr, left_child, start, mid)
            self._build(arr, right_child, mid + 1, end)
            self.tree[node] = self.tree[left_child] + self.tree[right_child]

    def _query(self, node, start, end, l, r):
        if r < start or l > end:
            return 0
        if l <= start and end <= r:
            return self.tree[node]
        mid = (start + end) // 2
        left_child = 2 * node + 1
        right_child = 2 * node + 2
        left_sum = self._query(left_child, start, mid, l, r)
        right_sum = self._query(right_child, mid + 1, end, l, r)
        return left_sum + right_sum

    def query(self, l, r):
        return self._query(0, 0, self.n - 1, l - 1, r - 1)

N, Q = map(int, input().split())
arr = list(map(int, input().split()))
segment_tree = SegmentTree(arr)
for _ in range(Q):
    U, V = map(int, input().split())
    print(segment_tree.query(U, V))
```

### **12. Given 2 integers N,M(where N is the number of vertices and M is the number of edges) followed by 'M' lines containing 2 vertices U,V and weight W. Find the shortest path from 1-N.If there is no path print '-1'.**

Input Size : 1<=N, M<=100000

**Example:**

**INPUT**

5 6

1 2 2

2 5 5

2 3 4

1 4 1

4 3 3

3 5 1

**OUTPUT**

5

```python
import heapq
from collections import defaultdict
import sys

def shortest_path(n, m, edges):
    graph = defaultdict(list)
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))
        
    dist = [float('inf')] * (n + 1)
    dist[1] = 0

    pq = [(0, 1)]
    while pq:
        current_dist, current_vertex = heapq.heappop(pq)
        if current_dist > dist[current_vertex]:
            continue
        for neighbor, weight in graph[current_vertex]:
            new_dist = current_dist + weight
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                heapq.heappush(pq, (new_dist, neighbor))
    return dist[n] if dist[n] != float('inf') else -1

input_data = sys.stdin.read().strip().split("\n")
n, m = map(int, input_data[0].split())
edges = [tuple(map(int, line.split())) for line in input_data[1:]]
print(shortest_path(n, m, edges))
```

### **14. Given 2 numbers N and Q followed by N numbers. Now Q inputs of 2 Numbers U and V are given. U indicates the starting index and V indicates the ending index. So for each U,V find the XOR of all values of the array from the index U to V (1 based indexing).**

HINT : USE SEGMENT TREE

Input Size : N<=100000

**Example:**

**INPUT**

5 3

1 2 1 4 1

1 3

2 4

3 4

**OUTPUT**

2

7

5

```python
class SegmentTree:
    def __init__(self, arr):
        n = len(arr)
        self.n = n
        self.tree = [0] * (4 * n)  # Allocate space for segment tree
        self._build(arr, 0, 0, n - 1)  # Build the tree
    
    def _build(self, arr, node, start, end):
        if start == end:
            # Leaf node
            self.tree[node] = arr[start]
        else:
            mid = (start + end) // 2
            left_child = 2 * node + 1
            right_child = 2 * node + 2
            # Recursively build the left and right children
            self._build(arr, left_child, start, mid)
            self._build(arr, right_child, mid + 1, end)
            # Internal node is the XOR of its children
            self.tree[node] = self.tree[left_child] ^ self.tree[right_child]

    def _query(self, node, start, end, l, r):
        if r < start or l > end:
            # Range outside the current segment
            return 0
        if l <= start and end <= r:
            # Current segment is fully within the range
            return self.tree[node]
        # Partial overlap: Query both children
        mid = (start + end) // 2
        left_child = 2 * node + 1
        right_child = 2 * node + 2
        left_xor = self._query(left_child, start, mid, l, r)
        right_xor = self._query(right_child, mid + 1, end, l, r)
        return left_xor ^ right_xor

    def query(self, l, r):
        # External query function (convert to 0-based indexing)
        return self._query(0, 0, self.n - 1, l - 1, r - 1)

N, Q = map(int, input().split())
arr = list(map(int, input().split()))

segment_tree = SegmentTree(arr)
for _ in range(Q):
    U, V = map(int, input().split())
    print(segment_tree.query(U, V))
```

### **15. Given a Binary Tree with N nodes and N-1 edges, print the mirror image of the tree with 1 as the root.**

Input Size : 1 <= N <= 100000

**Sample Testcases:**

**INPUT**

1

**OUTPUT**

1

```python
from collections import defaultdict, deque

def mirror_binary_tree(n, edges):
    # Build the adjacency list
    tree = defaultdict(list)
    for u, v in edges:
        tree[u].append(v)

    # Perform BFS to build the mirrored tree
    mirrored_tree = defaultdict(list)
    queue = deque([1])  # Start with root node (1 is always the root)
    visited = set()

    while queue:
        node = queue.popleft()
        visited.add(node)
        # Get children of the current node
        children = sorted(tree[node], reverse=True)  # Reverse to create a mirror
        for child in children:
            if child not in visited:
                mirrored_tree[node].append(child)
                queue.append(child)

    # Print the mirrored tree in BFS order
    result = []
    queue = deque([1])  # Start with root node
    while queue:
        node = queue.popleft()
        result.append(str(node))
        for child in mirrored_tree[node]:
            queue.append(child)
    return " ".join(result)

n = int(input())
edges = []
for _ in range(n - 1):
    u, v = map(int, input().split())
    edges.append((u, v))
print(mirror_binary_tree(n, edges))
```

### **17. A tree is a connected acyclic graph. A bipartite graph is a graph, whose vertices can be partitioned into 2 sets in such a way, that for each edge (U, V) that belongs to the graph, U and V belong todifferent sets. You can find more formal definitions of a tree and a bipartite graph in the notessection below. Consider a tree consisting of n nodes and add edges to it in such a way, that the graph is stillbipartite. Besides, after adding these edges the graph should be simple (should’nt contain loops ormultiple edges). What is the maximum number of edges they can add? A loop is an edge, which connects a node with itself. Graph doesn't contain multiple edges when foreach pair of nodes there is no more than one edge between them. A cycle and a loop aren't the same.**

Input Size : N<=100000, U,V<=N

**Example:**

**Input**

3

1 2

2 3

**Output**

0

```python
from collections import deque

def max_edges_in_bipartite_tree(n, edges):
    # Build the adjacency list
    tree = [[] for _ in range(n+1)]
    for u, v in edges:
        tree[u].append(v)
        tree[v].append(u)

    # Partition the tree nodes into two sets using BFS or DFS
    set1 = set()
    set2 = set()

    # BFS to split into two sets
    color = [-1] * (n + 1)  # -1: uncolored, 0: set1, 1: set2
    queue = deque([1])
    color[1] = 0  # Start with coloring the root as set1
    set1.add(1)

    while queue:
        node = queue.popleft()
        current_color = color[node]
        next_color = 1 - current_color
        for neighbor in tree[node]:
            if color[neighbor] == -1:
                color[neighbor] = next_color
                queue.append(neighbor)
                if next_color == 0:
                    set1.add(neighbor)
                else:
                    set2.add(neighbor)
    
    # Calculate the maximum number of edges
    a = len(set1)
    b = len(set2)

    # Maximum edges in a complete bipartite graph: a * b
    max_possible_edges = a * b

    # Subtract the existing edges in the tree: N-1
    max_additional_edges = max_possible_edges - (n - 1)
    return max_additional_edges

n = int(input())
edges = []
for _ in range(n - 1):
    u, v = map(int, input().split())
    edges.append((u, v))
print(max_edges_in_bipartite_tree(n, edges))
```

### **18. Given a Tree with N nodes and N-1 edges where each edge is given in a line in the form (u v), Print the level order representation of the tree in alternate manner(levels 1,3,5..). Consider 1 to be the root.**

Input Size : 2 <= N <= 100000

**Sample Testcases:**

**INPUT**

5

1 2

2 3

3 4

4 5

**OUTPUT**

1 3 5

```python
from collections import deque, defaultdict

def level_order_alternate(n, edges):
    # Build the adjacency list
    tree = defaultdict(list)
    for u, v in edges:
        tree[u].append(v)
        tree[v].append(u)

    # Initialize BFS
    queue = deque([1])  # Start BFS from node 1
    level = [-1] * (n + 1)  # To store the level of each node
    level[1] = 0  # Root node is at level 0
    result = []

    # Perform BFS
    while queue:
        node = queue.popleft()
        current_level = level[node]
        if current_level % 2 == 0:  # Collect nodes at odd levels (1, 3, 5, ...)
            result.append(node)

        for neighbor in tree[node]:
            if level[neighbor] == -1:
                level[neighbor] = current_level + 1
                queue.append(neighbor)    
    return result

n = int(input())
edges = []
for _ in range(n - 1):
    u, v = map(int, input().split())
    edges.append((u, v))
result = level_order_alternate(n, edges)
print(" ".join(map(str, result)))
```

```python
class SegmentTree:
    def __init__(self, arr):
        self.n = len(arr)
        self.tree = [0] * (4 * self.n)  # Segment tree array
        self.build(arr, 0, 0, self.n - 1)  # Build the tree

    def build(self, arr, node, start, end):
        if start == end:
            self.tree[node] = arr[start]  # Leaf node will have the value of the array
        else:
            mid = (start + end) // 2
            left_child = 2 * node + 1
            right_child = 2 * node + 2
            # Recursively build the segment tree
            self.build(arr, left_child, start, mid)
            self.build(arr, right_child, mid + 1, end)
            # Internal node will store the minimum of the two children
            self.tree[node] = min(self.tree[left_child], self.tree[right_child])

    def query(self, node, start, end, L, R):
        if R < start or end < L:
            return float('inf')  # Out of range, return infinity
        if L <= start and end <= R:
            return self.tree[node]  # Current segment is fully within range
        mid = (start + end) // 2
        left_child = 2 * node + 1
        right_child = 2 * node + 2
        # Partially in range, query the left and right children
        left_result = self.query(left_child, start, mid, L, R)
        right_result = self.query(right_child, mid + 1, end, L, R)
        return min(left_result, right_result)

def main():
    N, Q = map(int, input().split())
    arr = list(map(int, input().split()))
    
    # Segment Tree initialization
    seg_tree = SegmentTree(arr)
    
    # Process each query
    for _ in range(Q):
        U, V = map(int, input().split())
        # Query the segment tree, converting from 1-based to 0-based indexing
        result = seg_tree.query(0, 0, N - 1, U - 1, V - 1)
        print(result)

if __name__ == "__main__":
    main()
```
